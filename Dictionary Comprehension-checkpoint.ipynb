```python
# Function to count vowels in a word
def count_vowels(word):
    vowels = "aeiouAEIOU"
    return sum(1 for ch in word if ch in vowels)
words = ["apple", "banana", "cherry", "grape", "mango"]

vowel_count_dict = {word: count_vowels(word) for word in words}

print("Words List:", words)
print("Vowel Count Dictionary:", vowel_count_dict)
```
