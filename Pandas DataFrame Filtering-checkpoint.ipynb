```python
import pandas as pd

# Sample employee dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah'],
    'Department': ['HR', 'Finance', 'IT', 'IT', 'Finance', 'HR', 'Marketing', 'Finance'],
    'Salary': [45000, 55000, 70000, 72000, 60000, 48000, 65000, 80000]
}

# Create DataFrame
df = pd.DataFrame(data)
print("Employee DataFrame:\n", df, "\n")

# Calculate 75th percentile salary
percentile_75 = df['Salary'].quantile(0.75)
print("75th Percentile Salary:", percentile_75, "\n")

# Filter employees with salary above 75th percentile
high_earners = df[df['Salary'] > percentile_75]
print("Employees with Salary above 75th Percentile:\n", high_earners)
```
