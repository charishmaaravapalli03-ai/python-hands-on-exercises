```python
import pandas as pd

# Input Structure: Dictionary of students and their scores
students_scores = {
    'Alice': [85, 90, 88],
    'Bob': [70, 75, 80],
    'Charlie': [92, 88, 95],
    'David': [60, 65, 70]
}

# Step 1: DataFrame Creation 
# Convert the dictionary to a Pandas DataFrame
df = pd.DataFrame(students_scores)

# Step 2: Transpose the DataFrame 
df = df.T

# Rename columns for clarity
df.columns = ['Test1', 'Test2', 'Test3']

# Step 3: Calculate Average Score per Student 
df['Average_Score'] = df.mean(axis=1)

# Step 4: Display Final DataFrame 
print(df)
```
