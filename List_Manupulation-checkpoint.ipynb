```python
# Function to filter even numbers and sort them in descending order
def get_even_numbers_desc(numbers):
    
    even_nums = [num for num in numbers if num % 2 == 0]
    
    even_nums.sort(reverse=True)
    
    return even_nums

numbers = [12, 5, 7, 20, 33, 8, 14, 3]
result = get_even_numbers_desc(numbers)

print("Original List:", numbers)
print("Even Numbers (Descending):", result)
```
