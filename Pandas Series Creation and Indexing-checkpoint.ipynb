```python
import pandas as pd

# Part A: Create a Pandas Series from a list of Celsius temperatures
celsius = [0, 10, 20, 30, 40]
celsius_series = pd.Series(celsius, index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])
print("Celsius Series:\n", celsius_series)

# Part B: Convert Celsius to Fahrenheit and create a new Series with the same index
fahrenheit_series = celsius_series * 9/5 + 32
print("\nFahrenheit Series:\n", fahrenheit_series)
```

```python
Synopsis

Pandas Series Creation and Indexing – Temperature Conversion

Objective:
To create a Pandas Series representing temperatures in Celsius and convert them to Fahrenheit while maintaining consistent indexing.

Description:
This program demonstrates how to create and manipulate Pandas Series, a one-dimensional labeled array structure in Python.
First, a Series of temperature values in Celsius is created with day names as indices.
Next, a formula is applied to convert each Celsius value into Fahrenheit, resulting in another Series with the same index.
°F=(°C×59​)+32
Algorithm Steps:

Start the program.

Import the Pandas library.

Create a list of temperatures in Celsius.

Create a Pandas Series from the list and label the indices with weekdays.

Apply the Fahrenheit conversion formula to each element.

Store the converted values in a new Series with the same index.

Display both Series (Celsius and Fahrenheit).

End the program.
```
