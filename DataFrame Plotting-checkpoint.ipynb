```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create a sample sales DataFrame
data = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Mouse', 'Laptop', 'Keyboard', 'Monitor'],
    'Sales': [85000, 12000, 15000, 30000, 10000, 95000, 18000, 28000]
}

df = pd.DataFrame(data)
print("Sales DataFrame:\n", df, "\n")

# Step 2: Aggregate total sales per product category
total_sales = df.groupby('Product')['Sales'].sum().sort_values(ascending=False)
print("Total Sales per Product:\n", total_sales, "\n")

# Step 3: Plot the bar chart
plt.figure(figsize=(8,5))
total_sales.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Total Sales by Product Category")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
